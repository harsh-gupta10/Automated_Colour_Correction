```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.regularizers import l1, l2
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.layers import Dropout
from sklearn.model_selection import train_test_split, RandomizedSearchCV, KFold, cross_val_score
from tensorflow.keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv('./CSV_ABC/adobe5k_a.csv')
df
```

```python
X = df[["Raw Hue",'Raw Exposure','Raw Saturation']]
# X = X[0:4000]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# y = df['Edited Hue']
```

```python
# y = df['Edited Saturation']
```

```python
y = df['Edited Exposure']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanAbsoluteError


model = Sequential()
model.add(Dense(256, activation='elu', kernel_regularizer=l2(0.01))),
Dropout(0.3),
model.add(Dense(256, activation='elu', kernel_regularizer=l2(0.01))),
Dropout(0.3),
model.add(Dense(128, activation='elu', kernel_regularizer=l2(0.01))),
Dropout(0.2),
model.add(Dense(1, activation='linear'))  # Linear activation for regression
adam=Adam(learning_rate=0.001)
model.compile(optimizer=adam, loss='mae',metrics=['accuracy'])
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=1000, callbacks=[early_stopping], batch_size=100)
```

```python
model.evaluate(X_test, y_test)
```
